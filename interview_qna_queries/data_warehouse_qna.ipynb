**Q. what is difference between data modelling and dimensional modelling?**
- Data model is essential for organizing, storing, and retieving data efficiently in a way such as per business need.

**Why create a data model:**
- organize data logically:
	--Ensures data is structured in a meaningful way(tables, relationships, keys).
	--Makes the data easier to understand, manage, and query.
Ensure data quality and consistency:
	--Prevents data anomalies(Ex. Duplicates, inconsistencies...)
Improve Performance:
	--a good models helps with fast retrieval, especially for large datasets.
Enable scalability and maintenance:
	--Easier to adapt to new business requirements.
	--Modular and flexible design supports growth.
Align with business requirements.
	--Translates business processes into technical representations.


Data modelling:
--To organize data in a way that supports transactional systesm(OLTP).
--Organizing data for operations.

Key Characteristics:
--Focuses on normalization.
--Used to ensure data integrity.
--Entities represent real-world objects. Ex: Customer, Order, Product...
--Used to capture real-time transactions.

Data modeling useful for:
--Data governance.
--Data quality and auditing.
--Support data lineage.
--Matches source system structure.
--Reduces data redundancy.

Dimensional modeling:OLAP
--To structure data in a way that supports analytcal queries and reporting OLAP.
--Organizing data for analysis.
--Dimensional model is need in data warehouses and analytical systems.
--its make data easy to understand , fast to query, and efficient to report as per business users.

why create dimensional model:
Simplify complex data for business users:
	--Easier for non-technical users to explore and analyze data using BI tools.
	--Data is organized into facts and dimensions.
Enable fast and effient querying:
	--Dimensional models are optimized for OLAP queries, aggregations, and drill downs.
Improve performance in data warehouses:
	--Fewer joins.

Key Characteristics:
--Focuses on denormalization for performance.
--Uses star schema or snowflake schema.
--Designed for fast querying and aggregation.
--Contains 2 main components:
	--Fact: Quantitative data.Ex: Sales amount, profit,expenses....
	--Dimensions: Descriptive context. Ex: Date, Customer, Product, Country, Market..

Dimensional modeling useful:
--Fast, User-Friendly analytics.
--Supporting drill-downs, pivots and aggregations.

Note:
In Snowflake, we use data modelling(normalized) in staging/integration layers, and Dimensional modeling(denormalized) in presentation or reporting layers.
This approach gives a clean data pipelines with fast, intuitive reporting.

Staging layer:
--Cleaned, standardized data with minimal transformtion.

Integration/Transformation/ODS layer:
--Data is joined, deduplicated, and enriched, preserving business logic.

Presentation/Reporting Layer:Dimensional Modelling.
--Star/snowflake schemas for fast querying and analysis.

Note:
--We implement normally star schema as dimensional model in snowflake data warehouse.

Because:
--It is simple to query.
--Easy to understand for business users and analysts.

Note:
--Sometimes Snowflake schema modelling used.
--This is like star schema, but dimensions are more normalized.
--Dimension tables are split into sub-dimensions.
--while querying we need to use complex joins compared to star schema.

Steps to build a data model:
Requirements Gathering:
	--Understand what data is needed, how can be used, whata kind of problems can be solved.
	--We get business rules, data sources, use cases.
Conceptual data model:
	--High level, business focused model.
	--Focuses on entities like customer, order, product...
	--ER Diagram.
Logical data model:
	--here we will add more details like attributes, primary/foreign keys, data types...
	--Normalization is applied(1NF,2NF,3NF..).
Physical data model: SQL DDL Scripts.
	--Translate the logical model into a specific database platform like Snowflake, Oracle...
	--Column data types, constraints, indexes,partitioning, clustering in snowflake.
Build and Test:
	--populate data into tables.
	--Test the models like working as per business need.
Maintain and evolve:
	--As business needs change, update the model accordingly.
	
	
steps to create dimensional modelling:
Identify the business process:
	--like process sales transactions, customer support cases, inventory movements...etc.
Identify the grain(level of details):
	--Grain must be consistent, rows in the fact table represents.
	--Ex: one record per item sold in a sales transaction.
Identify the Dimensions:
	--How business do slice and dice on data?
	--Who, what, when, where, how.
	--Ex: Customer, Product, Time, Region, Salesperson...
Identify the Facts:Measures
	--These are the numeric values to be aggregated and analyzed.
	--Stored in the fact table.
	--Ex:Sales_amount, units_sold, discount, profit.
Design the schema:
	--use star or snowflake schema.
